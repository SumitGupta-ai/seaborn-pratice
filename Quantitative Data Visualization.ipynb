```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##### we use distribution plot to visualzie the distribution of quantitative data

##### let's discuss some plot that allow us to visualize the distribution of a data set. These plot are:

#### distplot/histoplot
#### jointplot
#### pairplot
#### rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
plt.figure(figsize=(5,5))
sns.histplot(df['total_bill'], bins=20, kde=True, color='r')
plt.title('Distribution of Total Bill', fontsize=10)
plt.show()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=20, kde=True, color='r')
plt.title('Distribution of Total Bill', fontsize=10)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins= 20, kde=True)
plt.show()
```

#### Jointplot

```python
# shows relationship between two numeric variable
sns.jointplot(x = 'total_bill', y= 'tip', data=df, kind='scatter', color='r', height=4., space=0.3,
                     marginal_kws= dict(bins = 25, fill=True, color='lightblue'))
plt.show()  
```

```python
# kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data=df, kind='kde', color='skyblue',
                 height=5)
plt.show() 
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data=df, kind='reg', color='g',
                     height= 4.7, space=0.3, marginal_kws= dict(bins = 25, fill=True, color='lightblue'))
plt.suptitle('Relationship Between Total Bill and Tip', fontsize=14, y=1.02)
plt.show() 
```

### pairplot

```python
# Shows relationship between multiple numeric variable.
sns.pairplot(df, hue='sex', palette='coolwarm',
             height=1.5)         
plt.suptitle("Pairwise Relationship Between Features in Tips Dataset",
             fontsize=10, y=1.04)
plt.show()
```

#### rugplot

```python
sns.rugplot(df, height=0.1)
plt.title('Total bill and tip & size', fontsize=10)
```

```python
plt.figure(figsize=(6,6))
sns.kdeplot(x = 'tip', data=df, fill=True,color='g')
plt.title('KDE Plot of Tip', fontsize=)
plt.show()
```
