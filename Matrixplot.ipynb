```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
tips = tips_df[['total_bill', 'tip', 'size']]
tips.head()
```

```python
tip_corr = tips.corr()
tip_corr
```

### HeatMap :-

##### A heatmap is used to visualize a matrix of data mostly correlation matrices using color to represent values.

```python
sns.heatmap(tip_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```

### Summary
#### A heatmap user color to show the strengh of relationships between numerical variable.

#### Blue = weak or negative relationship
#### Diagonal = same vaiable always 1 it helps to easily visualize how strongly features are related.

### Cluster map :-

#### A Clustermap is similar to a heatmap but it automatically groups clusters similar rows and columns together using hierarchical clustering

```python
sns.clustermap(tip_corr, annot=True,cmap='rainbow',
                  figsize=(8,6), linewidth=1)
plt.title('Cluster Map of Correlatiion Matrix', fontsize=10)
plt.show()
```

### ---Summary---

##### Clustermap is an advanced form of heatmap that group similar rows and columns together using hierarchical clustering.
##### it helps to find patterns and feature relationships in large datasets. Dendrogra, shows which variables are most similar.
