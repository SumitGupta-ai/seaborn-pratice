```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Categorical data plots

##### Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

##### 1. Countplot
##### 2. Barplot
##### 3. Stripplot
##### 4. Violinplot
##### 5. Stripplot

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df['day'].unique()
```

#### Countplot

```python
plt.subplot(1,2,1)                                                              # ---summary---
sns.countplot(x = 'day', hue='sex', data=df, palette='pastel',edgecolor='black')  #Creat subplot and customize.
plt.title('Count of Male vs Female Customer per day', fontsize=8)                 # palette= use- fill the theme
plt.subplot(1,2,2)                                                                # hue= use for category comparison.
sns.countplot(x = 'sex', hue='smoker', data=df,)
plt.title('count of Gender use smoker ', fontsize=8)
plt.show()
```

#### Barplot

```python
plt.figure(figsize=(6,4))
sns.barplot(x= 'day', y= 'total_bill', data=df, hue='sex', palette= 'coolwarm', estimator=np.sum,
               edgecolor='black', errorbar=None, )     # ---summary---
plt.title('Total Bill Per Day (by Gender)')            # hue= use second category for comparison                                           
plt.show()                                             # estimator: function for aggregation (mean median etc.) 
```

### Boxplot :-

##### A boxplot also called a box and whisker plot is used to visualize the spread and distribution of numerical data across different categories

```python
sns.boxplot(x = 'tip',  y= 'day', hue='day', data=df, palette= 'Set3',legend=True
               ,flierprops=dict(marker='o', markerfacecolor='r', markersize=8))
plt.title('Boxplot of Tip Per Day')
plt.show()
```

## Violinplot :-
##### A Violin plot shows the distribution probability denstiy median, and quartiles of numerical data for each category.


```python
sns.violinplot(x= 'time', y= 'tip', data=df, hue='sex', palette='Set2')
plt.title('Violin Plot Of Tip By Time (Lunch vs Dinner)')
plt.show()
```

## Stripplot

#### A strip plot displays all individual observations for a given categorical variable as dots on the graph.

```python
sns.stripplot(x = 'day', y = 'total_bill', hue='sex', data=df, jitter=True,
                 palette= 'rainbow')
plt.title('Strip plot of Total Bill by Gender and Day')
plt.show()
```
